## 3. Variables númericas

```python
print("Esatadísticas de 'Current Rank'\n")
print(f"Media: {df['Current Rank'].mean():.2f}")
print(f"Mediana: {df['Current Rank'].median():.2f}")
print(f"Desviación estándar: {df['Current Rank'].std():.2f}")
print(f"Varianza: {df['Current Rank'].var():.2f}")
print(f"Mínimo: {df['Current Rank'].min():.2f}")
print(f"Máximo: {df['Current Rank'].max():.2f}")
```

La media del Current Rank es 5.45, lo que indica que en promedio los deportistas se encuentran alrededor de la quinta posición del ranking. La mediana es 5, lo que confirma ya que no esta muy lejos de lo que nos dio el promedio.
  
La desviación estándar de 2.85 muestra que las posiciones no varían demasiado dentro del ranking, lo cual es lógico ya que el listado está limitado entre el puesto 1 y el 10. El valor mínimo corresponde al primer lugar, mientras que el máximo es el puesto 10.

```python
print("Esatadísticas de 'Year'\n")
print(f"Media: {df['Year'].mean():.2f}")
print(f"Mediana: {df['Year'].median():.2f}")
print(f"Desviación estándar: {df['Year'].std():.2f}")
print(f"Varianza: {df['Year'].var():.2f}")
print(f"Mínimo: {df['Year'].min()}")
print(f"Máximo: {df['Year'].max()}")
```

La media del año es 2005.12, lo que indica el punto medio de los datos. La mediana también es 2005 confirmando lo anterior siendo practicamente igual.

La desviación estándar de 9.06 años indica que los datos abarcan un período bastante amplio.

```python
print("Esatadísticas de 'earnings ($ million)'\n")
print(f"Media: {df['earnings ($ million)'].mean():.2f}")
print(f"Mediana: {df['earnings ($ million)'].median():.2f}")
print(f"Desviación estándar: {df['earnings ($ million)'].std():.2f}")
print(f"Varianza: {df['earnings ($ million)'].var():.2f}")
print(f"Mínimo: {df['earnings ($ million)'].min()}")
print(f"Máximo: {df['earnings ($ million)'].max()}")
```

La media de ganancias es de 45.52 millones de dólares, lo que representa el ingreso promedio anual de los deportistas, en el caso de la mediana es 39 millones, lo que nos indica que hay valores altos que estan inflando la media.
  
La desviación estándar de 33.53 muestra que existe una gran variabilidad en las ganancias entre los deportistas. El rango es amplio, con valores que van desde 8.1 millones hasta 300 millones de dólares, esto nos indica que hay mucha diferencia entre lo que ganan los diferentes deportistas.
