De esta categoría solo hice la moda ya que aunque tenia valores numéricos también tenía textos o datos como >30, siendo de momento dificil sacar estadisticas basicas de aca ya que habria que limpiar o transformar algunos datos como los NaN o >30.

## Variables númericas

```python
print(f"Media: {df['Current Rank'].mean():.2f}")
print(f"Mediana: {df['Current Rank'].median():.2f}")
print(f"Moda: {df['Current Rank'].mode()[0]:.2f}")
print(f"Desviación estándar: {df['Current Rank'].std():.2f}")
print(f"Varianza: {df['Current Rank'].var():.2f}")
print(f"Rango: {df['Current Rank'].max() - df['Current Rank'].min():.2f}")
print(f"Mínimo: {df['Current Rank'].min():.2f}")
print(f"Máximo: {df['Current Rank'].max():.2f}")
```
